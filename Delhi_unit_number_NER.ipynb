```python
import re

def extract_cts_code1(text):

    CTS_CODE_REGEX = r"(?<=NO).*"
    match = re.search(CTS_CODE_REGEX, text)
    if match:
        return match.group()
    else:
        return text

# Example usage:
#Corporation: Mumbai , Other details: Apartment/Flat No:602, Floor No:6TH FLOOR, C WING, Building Name:DEEPANKAR CHS, Block Sector:PRABHADEVI - 400025, Road:S L MATKAR MARG, City:Mahim, District:Mumbai,     C.T.S. Number : 295,  Leave and License Months:11
text = "NO 282-B"
text2 = "282-B"

# Extract the CTS code
cts_code = extract_cts_code1(text)
cts_code2 = extract_cts_code1(text2)

# Print the CTS code to the console
print(cts_code)
print(cts_code2)
```

```python
data["Final_Unit_no"] = data["Final_Unit_no"].apply(extract_cts_code1)
```

```python
import json

# Convert the data to JSON string
df = data1.to_json(orient='records')

# Write the JSON string to a file
with open('df.json', 'w') as f:
    f.write(df)

# Open the file and pass it to the json.load() function
with open('df.json', 'r') as f:
    df = json.load(f)

print(df[0])
```

```python
training_data = {'classes' : ['Unit_Number'], 'annotations' : []}
for example in df:
    temp_dict = {}
    temp_dict['text'] = example['Property Address']
    temp_dict['entities'] = []

    start = example['x']
    end = example['y']
    label = example['tag_name'].upper()
    temp_dict['entities'].append((start, end, label))
    training_data['annotations'].append(temp_dict)

print(training_data['annotations'][0])
```

```python
import spacy
from spacy.tokens import DocBin
from tqdm import tqdm

nlp = spacy.blank("en") # load a new spacy model
doc_bin = DocBin() # create a DocBin object
```

```python
from spacy.util import filter_spans
for training_example  in tqdm(training_data['annotations']):
    text = training_example['text']
    #print(text)
    labels = training_example['entities']
    #print(labels)
    doc = nlp.make_doc(text)
    #print("hbjbfkujubgfubjbukfb")
    #print(doc)
    ents = []
    #print(doc)
    for start, end, label in labels:
        span = doc.char_span(start, end, label=label, alignment_mode="expand")
        # print(f"\n00{span}")

        if span is None:
            print("Skipping entity")
        else:
        #if span is not None:
            ents.append(span)
        #ents.append(span)
    #print(ents)
    #print(span)
    filtered_ents = filter_spans(ents)
    #print(filtered_ents)
    doc.ents = filtered_ents
    #print(doc)
    doc_bin.add(doc)

doc_bin.to_disk("training_data.spacy") # save the docbin object
```

```python
def replace_empty_lists(df, col):
  """Replaces empty lists in a Pandas DataFrame column with `NA`.

  Args:
    df: A Pandas DataFrame.
    col: The name of the column containing lists.

  Returns:
    A Pandas DataFrame with empty lists replaced with `NA`.
  """

  df[col] = df[col].apply(lambda x: 'NA' if x == [] else x)
  return df

df = replace_empty_lists(test_data_1, 'model_unit_number')
print(df)
```
